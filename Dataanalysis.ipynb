```python
class Employee:
    def __init__(self,salary):
        self.salary = salary
        
```

```python
class Manager(Employee):
    def department(self, dept_name):
        self.dept_name= dept_name
        return f"I work in the {dept_name} department and my salary is {self.salary}"
```

```python
worth =Manager(5000)
print(worth.department("HR"))
```
