```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats import contin
```

```python
coin_toss_counts = sb_data["Coin Toss Outcome"].value_counts()
```

```python
fair_coin_expected =[sum(coin_toss_counts)/2,sum(coin_toss_counts)/2]
fair_coin_expected
```

```python
fair_coin_observed = coin_toss_counts.values
fair_coin_observed
```

```python
fair_coin_result=stats.chisquare(fair_coin_observed,fair_coin_expected)
fair_coin_result
```

```python
# Data from FiveThirtyEight
nfl_games = pd.read_csv("nfl_games.csv")
# Narrow to relevant rows
nfl_games = nfl_games[(nfl_games["neutral"] == 0) & (nfl_games["playoff"] == 0) & (nfl_games["season"] >= 1966 )]
# Engineer a new feature
nfl_games.loc[nfl_games["result1"] == 0, "Game Winner"] = "Away Team"
nfl_games.loc[nfl_games["result1"] == 1, "Game Winner"] = "Home Team"

game_winner_counts_nfl = nfl_games["Game Winner"].value_counts().sort_index()
game_winner_counts_nfl
```

```python
nfl_observed = game_winner_counts_nfl.values
nfl_observed
```

```python
nfl_expected = [sum(game_winner_counts_nfl) / 2,sum(game_winner_counts_nfl) / 2]
nfl_expected
```

```python
nfl_chi_square = stats.chisquare(nfl_observed, nfl_expected)
nfl_chi_square
```
