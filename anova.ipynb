```python
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
```

```python
house_prices = pd.read_csv(r"house_prices_dataset.csv")
house_prices.head()
```

```python
house_prices.columns
```

```python
house_prices=house_prices.drop(columns=["zipcode","lat","long"],axis=1)
house_prices.head()
```

```python
# Correct formula syntax and categorical variable notation
formula = "price ~ bedrooms + bathrooms + sqft_living + sqft_lot + floors + year_built + year_renovated + C(waterfront) + C(view) + C(condition) + C(grade)"
lm = ols(formula=formula, data=house_prices).fit()
table = sm.stats.anova_lm(lm, typ=2)
print(table)
```
