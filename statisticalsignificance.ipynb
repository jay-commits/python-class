```python
#z-test 
from scipy import stats
import numpy as np
sample_size = 40
population_mean = 100
population_sd=16
alpha=0.05
x_bar=103
z_statistic = (x_bar - population_mean) / (population_sd / np.sqrt(sample_size))
z_statistic
```

```python
z_cdf=stats.norm.cdf(z_statistic)
```

```python
p_value=1-z_cdf
p_value
```

```python
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Do not reject the null hypothesis")
```
