```python
# Simple chatbot for beginners to answer data science and AI questions
def chatbot():
    # Dictionary with questions and answers
    answers = {
        "what is data science?": "Data science is using data to find answers and make decisions. It uses math, coding, and knowledge about a topic.",
        "what is ai?": "AI, or Artificial Intelligence, is when computers do smart things like learning or solving problems, like humans do.",
        "what is machine learning?": "Machine learning is a part of AI where computers learn from data to make predictions or choices.",
        "what is python?": "Python is a popular programming language used in data science because itâ€™s easy to learn and has tools for working with data."
    }

    # Welcome message
    print("Hi! I'm a Data Science & AI Chatbot. Ask me a question or type 'quit' to stop.")
    print("Example questions: 'what is data science?', 'what is ai?'")

    # Keep running until the user types 'quit'
    try:
        while True:
            # Get user's question
            question = input("Your question: ").lower().strip()

            # Check if user wants to quit
            if question == "quit":
                print("Chatbot: Bye! Thanks for asking questions.")
                break

            # Check if the question is in our answers
            if question in answers:
                print("Chatbot:", answers[question])
            else:
                print("Chatbot: Sorry, I don't know that one! Try asking 'what is data science?' or 'what is ai?'")
    except KeyboardInterrupt:
        print("\nChatbot: Stopped by user. Bye!")
    except Exception as e:
        print(f"Chatbot: An error occurred: {e}. Please try again.")

# Check if Python is running the script directly
if __name__ == "__main__":
    print("Starting the chatbot...")
    chatbot()
    
```
