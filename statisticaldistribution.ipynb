```python
pumpkin=["a", "b", "c", "d", "e", "f"]
weights=[19,14,15,9,10,17]
```

```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import itertools 
pumpkin_dict= dict(zip(pumpkin,weights))
pumpkin_dict
```

```python
#calculate population mean
def calculate_mu(x):
    d = sum(x.values()) / len(x)
    return d
```

```python
pumpkin_mu=calculate_mu(pumpkin_dict)
pumpkin_mu
```

```python
plt.bar(pumpkin, weights, color="orange", edgecolor="black")
plt.axhline(pumpkin_mu, color="red", linestyle="--", linewidth=2, label=f"Mean = {pumpkin_mu}")
plt.xlabel("Pumpkins")
plt.ylabel("Weight")
plt.title("Pumpkin Weights with Mean Highlighted")
plt.legend()
plt.show()
```

```python
sample_size= 2
combs = list(itertools.combinations(pumpkin_dict.keys(),sample_size))
combs
```

```python
def sample_mean(n,data):
    n=n#samplesize
    mu= sum(data.values())/len(data)
    #generate all possible combinations using the given sample size
    combinations = list(itertools.combinations(data.keys(),n))
    #calculate sample mean
    xbarlist=[]
    for comb in combinations:
        sample_weights=[data[p] for p in comb]
        sample_mean=sum(sample_weights)/n
        xbarlist.append(sample_mean)
    return xbarlist, combinations
```

```python
combinations,xbarlist= sample_mean(sample_size,pumpkin_dict)
for c,m in zip(combinations,xbarlist):
    print(c,m)
```

```python
mean_of_sample_means = sum(combinations) / len(xbarlist)
mean_of_sample_means
```

```python
# Lists of people and their heights
people = ["Alice", "Bob", "Charlie", "Diana", "Ethan", "Fiona", "George", 
          "Hannah", "Isaac", "Julia", "Kevin", "Laura", "Michael", "Nina", "Oscar"]
heights = [165, 178, 172, 160, 185, 170, 182, 158, 174, 169, 180, 162, 176, 168, 181]

# Create a dictionary (like a phonebook) to pair names with heights
height_dict = {name: height for name, height in zip(people, heights)}

# Function to find the average height from a random sample
def average_height(sample_size, data):
    import random
    # Pick random heights
    chosen_heights = random.sample(list(data.values()), sample_size)
    # Add them up and divide by how many we picked
    return sum(chosen_heights) / sample_size

# Calculate average for 3 people
result = average_height(3, height_dict)
print(f"Average height of 3 people: {result:.2f} cm")
```
